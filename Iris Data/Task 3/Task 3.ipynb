## Task 3: Basic Data Visualization

ðŸŸ¢ Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

ðŸŸ¢ Step 2: Load Cleaned Dataset




```python
df = pd.read_csv(r'/Users/sahilraza/Documents/Codveda _Internship/iris.csv', encoding='unicode_escape')
df.head()
```

ðŸ“Š 1. Bar Plot â€“ Count of Each Species



```python
sns.countplot(data=df, x='species', hue='species', palette='Set2', legend=False)
plt.title('Species Count')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

ðŸ“ˆ 2. Line Chart â€“ Average Sepal Length per Species

```python
avg_sepal = df.groupby('species')['sepal_length'].mean()
avg_sepal.plot(kind='line', marker='o', color='orange')
plt.title('Average Sepal Length per Species')
plt.xlabel('Species')
plt.ylabel('Avg Sepal Length')
plt.show()
```

âšª 3. Scatter Plot â€“ Sepal Width vs Petal Width

```python
sns.scatterplot(data=df, x='sepal_width', y='petal_width', hue='species', palette='Set1')
plt.title('Sepal Width vs Petal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.show()
```

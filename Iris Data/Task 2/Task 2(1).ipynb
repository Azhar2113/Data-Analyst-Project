## Task 2: Exploratory Data Analysis (EDA)

Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

 Step 2: Load Cleaned Dataset

```python
df = pd.read_csv(r'/Users/sahilraza/Documents/Codveda _Internship/iris.csv', encoding='unicode_escape')
df.head()
```

 Step 3: Summary Statistics

```python
# Mean, Median, Mode, Standard Deviation
print("Mean:\n", df.mean(numeric_only=True))
print("\nMedian:\n", df.median(numeric_only=True))
print("\nMode:\n", df.mode().iloc[0])
print("\nStandard Deviation:\n", df.std(numeric_only=True))
```

Step 4: Correlation Matrix

```python
# Correlation between numerical columns
correlation = df.corr(numeric_only=True)
print("\nCorrelation Matrix:\n", correlation)
```

ðŸŸ¢ Step 5: Visualizations

ðŸ“Š Histogram

```python
df['sepal_length'].plot(kind='hist', bins=15, color='skyblue', edgecolor='black', title='Sepal Length Distribution')
plt.xlabel("Sepal Length")
plt.show()
```

ðŸ“¦ Boxplot

```python
sns.boxplot(data=df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
plt.title("Boxplot of Numerical Features")
plt.show()
```

âšª Scatter Plot

```python
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title("Sepal Length vs Petal Length by Species")
plt.show()
```

ðŸ”¥ Heatmap (Correlation)

```python
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
